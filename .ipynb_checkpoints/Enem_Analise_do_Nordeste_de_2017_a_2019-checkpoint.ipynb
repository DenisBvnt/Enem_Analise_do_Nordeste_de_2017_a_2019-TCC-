```python
def countplot_without_hue(data, column, percentage_style=None):
  plot = sns.countplot(data=data, x=column, order=data[column].value_counts().index.to_list())
  sns.despine(bottom = True, left = True)
  if percentage_style == 'center_white':
    for p in plot.patches:
        height = p.get_height()
        plot.text(x=p.get_x()+p.get_width()/2., y=height-height/2, s='{:.2f}%'.format(100 * height/len(data)),
                  ha="center", color='white', fontweight='heavy', fontsize=16)
  elif percentage_style == 'top_black': 
    for p in plot.patches:
      height = p.get_height()
      plot.text(x=p.get_x()+p.get_width()/2., y=height + min(0.003*len(data), 10000), s='{:.2f}%'.format(100 * height/len(data)),
                ha="center", fontweight='heavy', fontsize=max([10, min([12, 48/data[column].nunique()])]))


def countplot_with_hue(data, x_column, hue_column, number_of_categories, hue_categories):
  plot = sns.countplot(data=data, x=x_column, hue=hue_column, order=data[x_column].value_counts().index.to_list(),
                       hue_order=data[hue_column].value_counts().index.to_list())
  sns.despine(bottom = True, left = True)
  a = [p.get_height() for p in plot.patches]
  patch = [p for p in plot.patches]
  for i in range(number_of_categories):
      total = data[x_column].value_counts().values[i]
      for j in range(hue_categories):
          percentage = '{:.2f}%'.format(100 * a[(j*number_of_categories + i)]/total)
          x = patch[(j*number_of_categories + i)].get_x() + patch[(j*number_of_categories + i)].get_width() / 5
          y = patch[(j*number_of_categories + i)].get_y() + patch[(j*number_of_categories + i)].get_height()
          plot.annotate(percentage, (x, y), size = 10, fontweight='heavy', fontsize=10)

def boxplot_without_hue(data, x, y):
  plot = sns.boxplot(data=data, x=x, y=y, order=data[x].value_counts().index.to_list())
  sns.despine(bottom = True, left = True)


def boxplot_one_axis(data, column, orient='h'):
  plt.subplots_adjust(wspace=0.3)
  if orient == 'h':
    sns.boxplot(data=data, x=column, width=.5)
    plt.tick_params(left=False, bottom=True, labelbottom=True, labelleft=False)
    plt.text(x=0.55*data[column].max(), y=-0.25, s='Média: {0:.2f}\nMediana: {1:.2f}\nDesvio padrão: {2:.2f}'
    .format(data[column].mean(), data[column].median(), data[column].std()), fontsize=12)
  elif orient == 'v':
    sns.boxplot(data=data, y=column)
    plt.tick_params(left=True, bottom=False, labelbottom=False, labelleft=True)
    plt.text(x=0.05, y=0.8*data[column].max(), s='Média: {0:.2f}\nMediana: {1:.2f}\nDesvio padrão: {2:.2f}'
    .format(data[column].mean(), data[column].median(), data[column].std()), fontsize=12)
  plt.title('{}'.format(int(data['NU_ANO'].mode()[0])), fontsize=20, color='grey', fontweight='heavy', pad=20)
  sns.despine(bottom = True, left = True)


def violin_box(data, x=None, y=None, orient='v'):
  if (x == None) | (y == None):
    ax = sns.boxplot(data=data, x=x, y=y, width=0.15, showfliers=False, showcaps=False)
    ax = sns.violinplot(data=data, x=x, y=y, inner=None)
  else:
    if orient == 'v':
      ax = sns.boxplot(data=data, x=x, y=y, width=0.3, showfliers=False, showcaps=False, order=data.groupby(x)[y].mean().sort_values().index)
      ax = sns.violinplot(data=data, x=x, y=y, inner=None, order=data.groupby(x)[y].mean().sort_values().index)
    elif orient == 'h':
      ax = sns.boxplot(data=data, x=x, y=y, width=0.3, showfliers=False, showcaps=False, order=data.groupby(y)[x].mean().sort_values().index)
      ax = sns.violinplot(data=data, x=x, y=y, inner=None, order=data.groupby(y)[x].mean().sort_values().index)

  plt.setp(ax.collections, alpha=.4)
  sns.despine(bottom = True, left = True)
```

###5.2.1.2. <font color='red'>Substituindo os outliers de idade</font>

```python
#Alunos menores de 14 anos que concluíram ou estão cursando o ensino médio ou
#alunos menores de 14 anos que não são treineiros
for ano in data_list:
  print('{} - {}'.format(ano['NU_ANO'].max(), ano[((ano['NU_IDADE'] < 14) & (ano['TP_ST_CONCLUSAO'] != 4)) | ((ano['NU_IDADE'] < 14) & (ano['IN_TREINEIRO'] == 0))].shape[0]))
```

Não existe uma idade mínima para fazer o Enem. Porém, se você estiver no segundo ano do ensino médio ou em alguma série anterior, terá que fazê-lo como treineiro e não poderá usar a sua nota para se candidatar a uma vaga em universidades, portanto, iremos modificar todos os registros menores que 14 anos e que ao mesmo tempo NÃO são treineiros ou menores de 14 anos que concluíram ou estão cursando o ensino médio.

Substituirémos a idade desses determinados participantes por um valor aleatório dentro do intervalo (média - desvio padrão, média + desvio padrão).

```python
for ano in data_list:
  mask = ((ano['NU_IDADE'] < 14) & (ano['TP_ST_CONCLUSAO'] != 4)) | ((ano['NU_IDADE'] < 14) & (ano['IN_TREINEIRO'] == 0))
  lim_minimo = ano['NU_IDADE'].mean() - ano['NU_IDADE'].std()
  lim_maximo = ano['NU_IDADE'].mean() + ano['NU_IDADE'].std()
  ano['NU_IDADE'].mask(mask, np.random.randint(lim_minimo, lim_maximo), inplace=True)
```

Quanto aos participantes com idades mais elevadas, não iremos tratá-los, pois como dito acima "Foi considerado treineiro o inscrito que não havia concluído o ensino médio e não o concluiria em 2019", logo, não podemos identificar quais deles são outliers, pois no Brasil há um grande quantidade de idosos que não possuem ensino médico completo.

####6.3.3.1. Classes sociais



Iremos criar uma variável onde transformaremos o valor de renda familiar mensal em classe social, levando me consideração a classificação do IBGE.

**Classificação do IBGE:**
* Classe A = mais que 20 salários mínimos
* Classe B = entre 10 e 20 salários mínimos
* Classe C = entre 4 e 10 salários mínimos
* Classe D = entre 2 e 4 salários mínimos
* Classe E = menor que 2 salários mínimos

</br>Para a criação da variável de classe social, utilizaremos a variável de renda mensal familiar e faremos a seguinte transformação:
- Q -> Classe A 
- N, O e P -> Classe B
- H, I, J, K, L e M -> Classe C
- E, F e G -> Classe D
- A, B, C e D -> Classe E 

```python
for i, ano in enumerate(data_list):
  ano['CLASSE_SOCIAL'] = ano['RENDA_MENSAL_FAMILIAR']
  ano['CLASSE_SOCIAL'].replace({'A':'E', 'B':'E', 'C':'E', 'D':'E', 'E':'D', 'F':'D', 'G':'D', 'H':'C', 'I':'C', 'J':'C',
                                'K':'C', 'L':'C', 'M':'C', 'N':'B', 'O':'B', 'P':'B', 'Q':'A'}, inplace=True)
```

```python
X['Cor'].replace({0:'N/D', 1:'Branca', 2:'Preta', 3:'Parda', 4:'Amarela', 5:'Indígena'}, inplace=True)
X['Sexo'].replace({'F':'Feminino', 'M':'Masculino'}, inplace=True)
X['Mae ensino'].replace({'A':'Fundamental 1 incompleto', 'B':'Fundamental 1 incompleto', 'C':'Fundamental 2 incompleto',
                         'D':'Fundamental 2 completo', 'E':'Medio completo', 'F':'Superior completo',
                         'G':'Superior completo', 'H':'Não sabe'}, inplace=True)
X['Pai ensino'].replace({'A':'Fundamental 1 incompleto', 'B':'Fundamental 1 incompleto', 'C':'Fundamental 2 incompleto',
                         'D':'Fundamental 2 completo', 'E':'Medio completo', 'F':'Superior completo',
                         'G':'Superior completo', 'H':'Não sabe'}, inplace=True)
X['Classe social'].replace({'A':'Classe B|A', 'B':'Classe B|A', 'C':'Classe C', 'D':'Classe D', 'E':'Classe E'}, inplace=True)
X['Computador'] = X['Computador'].apply(lambda x: 'Nao tem' if x == 'A' else 'Tem')
X['Celular'].replace({'A':'1 ou nenhum', 'B':'1 ou nenhum', 'C':'2 ou 3', 'D':'2 ou 3', 'E':'4 ou mais'}, inplace=True)
X['Internet'].replace({'A':'Nao tem', 'B':'Tem'}, inplace=True)
X['Cluster'] = X['Cluster'].astype(str)
X['Nota'] = pd.qcut(x=X['Nota'], q=3).astype(str).str.strip('()[]').apply(lambda x: x.split(', ')[0]+'-'+x.split(', ')[1])
X['Nota'].replace({sorted(X['Nota'].unique())[0]:'33% menores', sorted(X['Nota'].unique())[1]:'33% do meio',
                   sorted(X['Nota'].unique())[2]:'33% maiores'}, inplace=True)
```

```python
#Calcular a confiança de relações 1 pra 1
def return_conf(a, b):
  df = pd.DataFrame(columns=['X', 'Y', 'Conf'])
  for x in X[a].unique():
    for y in X[b].unique():
      cf = X[(X[a] == x) & (X[b] == y)].shape[0]/len(X[X[a] == x])
      df = df.append({'X':x, 'Y':y, 'Conf':cf}, ignore_index=True)
  return df
```

```python
#Criando DataFrame com as confianças das relações
confs = pd.DataFrame(columns=['X', 'Y', 'Conf'])
for x in X.columns:
  for y in X.drop(x, axis=1).columns:
    confs = pd.concat([confs, return_conf(x, y)], ignore_index=True)
```

```python
 confs = confs[confs['X'] != confs['Y']].sort_values('Conf', ascending=False)
```

```python
itemsets, regras = apriori(transactions, min_support=0.01, min_confidence=0.7, max_length=4, verbosity=1)
```

```python
lista_regras = []
lista_supp = []
lista_conf = []
lista_lift = []
lista_conv = []

for regra in regras:
    lista_regras.append(regra)
    lista_supp.append(round(regra.support * 100, 2))
    lista_conf.append(round(regra.confidence * 100, 2))
    lista_lift.append(round(regra.lift, 2))
    lista_conv.append(round(regra.conviction, 2))
```

```python
tabela_regras = pd.DataFrame({'Regra':lista_regras, 'Suporte (%)':lista_supp, 'Confiança (%)':lista_conf, 'Lift':lista_lift, 'Convicção':lista_conv})
```
